```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos desde un archivo CSV en un DataFrame de pandas
New_Data = pd.read_csv('acc_20.csv')

# Asegúrate de que los nombres de las columnas sean correctos. Puedes imprimir los nombres de las columnas para verificarlo.
print(New_Data.columns)

# Crear una figura con dos subtramas dispuestas en una fila
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Graficar el primer subtrama: Accident Severity by Day of Week
sns.lineplot(x='DAY_WEEKNAME', y='URBANICITYNAME', data=New_Data, ax=axes[0])  # Asegúrate de que 'DAY_WEEKNAME' y 'URBANICITYNAME' son correctos
axes[0].set_title('Accident Severity by Day of Week', fontsize=12)
axes[0].set_xlabel('Day of Week', fontsize=10)
axes[0].set_ylabel('Accident Severity', fontsize=10)
axes[0].tick_params(axis='x', rotation=45, labelsize=10)
axes[0].grid(True)

# Graficar el segundo subtrama: Accident Severity by Month
sns.lineplot(x='MONTHNAME', y='URBANICITYNAME', data=New_Data, ax=axes[1])  # Asegúrate de que 'MONTHNAME' y 'URBANICITYNAME' son correctos
axes[1].set_title('Accident Severity by Month', fontsize=12)
axes[1].set_xlabel('Month', fontsize=10)
axes[1].set_ylabel('Accident Severity', fontsize=10)
axes[1].tick_params(axis='x', rotation=45, labelsize=10)
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

Grafica circular

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el conjunto de datos
df = pd.read_csv('acc_20.csv')

# Asumiendo que quieres hacer un gráfico circular de la columna "CATEGORIA"
# Primero, necesitamos el conteo de cada categoría
conteo_categorias = df['MONTHNAME'].value_counts()

# Crear la gráfica circular
plt.figure(figsize=(10, 7)) # tamaño de la figura, puedes ajustarlo según tus necesidades
plt.pie(conteo_categorias, labels = conteo_categorias.index, autopct='%1.1f%%') # autopct para mostrar porcentajes
plt.title('Distribución de los meses') # cambia 'CATEGORIA' por el nombre real de tu columna
plt.show()
```
